```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import math

def plot(name, superposition):
    fig = plt.figure(figsize=(10,10))
    ax = fig.gca()
    
    if superposition:
        vectors = []
        for angle in range(0, 91, 5):
            vectors.append([0, 0, math.cos(math.radians(angle)), math.sin(math.radians(angle))])
        soa = np.array(vectors)
        X, Y, U, V = zip(*soa)
        ax.quiver(X, Y, U, V, angles='xy', headaxislength=7, headlength=7, headwidth=7, scale_units='xy', scale=1, color=["silver"])

    qubit_angle = math.pi / 4
    qubits = [[0, 0, math.cos(qubit_angle), math.sin(qubit_angle)]]
    q_projections = []
    for qubit in qubits:
        q_projections.append([qubit[2], 0, 0, qubit[3]])
        q_projections.append([0, 0, qubit[2], 0])

    q_X, q_Y, q_U, q_V = zip(*np.array(qubits))
    qp_X, qp_Y, qp_U, qp_V = zip(*np.array(q_projections))
    ax.quiver(q_X, q_Y, q_U, q_V, headaxislength=7, headlength=7, headwidth=7, scale_units='xy', angles='xy', scale=1, color=["green"])
    ax.quiver(qp_X, qp_Y, qp_U, qp_V, headwidth=0,headlength=0, headaxislength=0, color="tomato", scale=1, hatch="-")

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_aspect("equal")

    ax.set_xlabel("0 state", fontsize="xx-large")
    ax.set_ylabel("1 state", fontsize="xx-large")
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

    plt.draw()

    plt.savefig(name, facecolor='white', transparent=False, bbox_inches="tight")
    plt.show()
```

```python
plot("super", True)
```

```python
plot("single", False)
```
